```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
shows = pd.read_csv('netflix_titles.csv')
shows.head()
```

```python
shows.isnull().sum() # check for missing values
```

```python
sns.heatmap(shows.isnull(), yticklabels=False, cbar=False, cmap='viridis'); # visualize missing values with a heatmap
```

## EDA

```python
shows["rating"].value_counts().unique() # check for unique values in the rating column
```

```python
shows.drop(["director", "cast"], axis=1, inplace=True) # drop the director and cast columns
```

```python
shows.head()
```

```python
shows["country"].replace(np.nan, "United States", inplace=True) # replace missing values in the country column with United States
```

```python
shows.drop(["date_added"], axis=1, inplace=True) # drop the date_added column
```

```python
shows.head()
```

```python
shows["rating"].value_counts()
```

```python
shows["listed_in"].value_counts()
```

```python
shows["rating"].replace(np.nan, "TV-MA", inplace=True) # replace missing values in the rating column with TV-MA
```

```python
shows.isnull().sum() # check for missing values
```

## Data Visualization

```python
sns.countplot(x="type", data=shows); # visualize the count of movies and TV shows
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="rating", data=shows); # visualize the count of ratings
```

```python
plt.figure(figsize=(35, 6))
sns.countplot(x="release_year", data=shows); # visualize the count of release years
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x="rating", y="type", data=shows); # visualize the count of ratings
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="rating", data=shows, hue="type"); # visualize the count of ratings
```

```python
import plotly.express as px
top_countries = shows[0:10]
fig = px.sunburst(top_countries, path=['country'])
fig.show() # visualize the top 10 countries
```

```python
shows["rating"].value_counts().plot.pie(autopct="%1.1f%%", figsize=(20, 35)); # visualize the ratings as a pie chart
plt.show()
```

```python
country_counter = shows["country"].value_counts().sort_values(ascending=False) # count the number of shows per country
country_counter = pd.DataFrame(country_counter)
top_country = country_counter[0:11] # get the top 10 countries
top_country
```

```python
old = shows.sort_values("release_year", ascending=True) # sort the shows by release year
old = old[old["duration"] != ""]
old[["title", "release_year"]][:15]
```

```python
tag = "Stand-Up Comedy"
shows["relevant"] = shows["listed_in"].fillna("").apply(lambda x: 1 if tag.lower() in x.lower() else 0)
comm = shows[shows["relevant"] == 1]
comm[comm["country"] == "United States"][["title", "country", "release_year"]].head(10)
```

```python
tag = "Kids' TV" 
shows["relevant"] = shows["listed_in"].fillna("").apply(lambda x: 1 if tag.lower() in x.lower() else 0)
comm = shows[shows["relevant"] == 1]
comm[comm["country"] == "United States"][["title", "country", "release_year"]].head(10)
```

```python
shows_countries = pd.DataFrame(shows.country.value_counts().reset_index().values, columns=["country", "count"])
shows_countries.head()
```

```python
fig = px.choropleth(locationmode="country names", 
    locations=shows_countries.country, 
    labels=shows_countries["count"]
    )
fig.show() # visualize the number of shows per country
```

```python
data = pd.DataFrame(shows.release_year.value_counts().reset_index().values, columns=["year", "count"])
data.head()
```

```python
plt.figure(figsize=(16, 6))
shows[shows["type"] == "Movie"]["release_year"].value_counts()[:20].plot(kind="bar", color="Red"); # visualize the number of movies per year
plt.title(
    "Number of Movies per Year of Any Kind on Netflix",
)
```

```python
plt.figure(figsize=(16, 6))
shows[shows["type"] == "TV Show"]["release_year"].value_counts()[:20].plot(kind="bar", color="Blue"); # visualize the number of TV shows per year
plt.title("Number of TV Shows per Year of Any Kind on Netflix")
```

```python
plt.figure(figsize=(16, 10))
shows[shows["type"] == "Movie"]["listed_in"].value_counts()[:10].plot(kind="barh", color="Black"); # visualize the number of movies per genre
plt.title("Top 10 Genres of Movies on Netflix", size=20)
```

```python
plt.figure(figsize=(12, 6))
shows[shows["type"] == "TV Show"]["listed_in"].value_counts()[:10].plot(kind="barh", color="Green"); # visualize the number of TV shows per genre
plt.title("Top 10 Genres of TV Shows on Netflix", size=20)
```
